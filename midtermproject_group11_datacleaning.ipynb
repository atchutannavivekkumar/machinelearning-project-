In the cleaning process, we noticed that the age column has decimal values. Which cannot be happended. so we converted them into integer values. 

```python
obesity_dataset_begin['Age'] = obesity_dataset_begin['Age'].astype(int)
```

The data set has two columns height and weight, which can be drop by getting bmi as value

```python
obesity_dataset_begin['Weight'] = pd.to_numeric(obesity_dataset_begin[' Weight'], errors='coerce')
obesity_dataset_begin['Height'] = pd.to_numeric(obesity_dataset_begin[' Height'], errors='coerce')

obesity_dataset_begin['Bmi_index'] = obesity_dataset_begin['Weight'] / (obesity_dataset_begin['Height'] ** 2)
obesity_dataset_begin.drop([' Weight', ' Height'], axis=1, inplace=True)
```

```python
obesity_dataset_begin = obesity_dataset_begin.drop(['Weight'], axis=1)
obesity_dataset_begin = obesity_dataset_begin.drop(['Height'], axis=1)
```

```python
obesity_dataset_begin['Bmi_index'].fillna(obesity_dataset_begin['Bmi_index'].mean(), inplace=True)
```
